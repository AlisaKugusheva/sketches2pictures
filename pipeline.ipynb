## Pix2Pix Pipeline 

### Input

* `imgs` - list with `str` paths to the source images
* `masks` - list with `str` paths to the masks obtained after segmentation from the corresponding image (i.e. `imgs[0]` has mask at `masks[0]`)

### Steps

1. Creates dataset which takes given `imgs` and `masks` at initialisation and return images without background.

2. Splits the dataset into two folders: train and test. 

3. Applies `save_img_and_masks` function to each part. Gets combined source image and its sketch for train dataset and combined empty white image and sketch for test dataset (because we still need to save the combined format). Write results on a disk.

4. Trains GAN pix2pix model on the train sample.

5. Uses trained model on the test sample.

### Data Preprocessing Helpers

```python
class DatasetWithMasks(Dataset) :
    def __init__(self, imgs, masks):
        self.imgs = imgs
        self.masks = masks
    
    def __len__(self) :
        return len(self.imgs)
    
    def __getitem__(self, idx) :
        img = cv2.imread(self.imgs[idx])
        mask = np.array(Image.open(self.masks[idx]))
        img[mask == 0] = 255
        
        return img
```

```python
def transform(img, size: tuple, crop_size: tuple):
    """
    returns resized and center cropped image
    arguments:
        -img: original image
        -size: size of image after applying resize
        -crop_size: size of image after applying center crop
    """
    img_resized = cv2.resize(img, size)
    w, h = crop_size
    center = img_resized.shape [0] / 2, img_resized.shape [1] / 2
    x = center[1] - w/2
    y = center[0] - h/2
    crop_img = img_resized[int(y):int(y+h), int(x):int(x+w)]
    return crop_img

def write_combined(dataset_path: str, index: int, img_combined):
    try:
        os.makedirs(dataset_path)
    except FileExistsError:
    # directory already exists
        pass
    # save image and its sketch combined in one file
    cv2.imwrite(dataset_path+str(index)+'.jpg', img_combined)

def save_img_and_masks(img, 
                       dataset_path: str, 
                       index: int, 
                       size: tuple,
                       crop_size: tuple,
                       canny_threshold_1: int,
                       canny_threshold_2: int):
    """
    This function gets an image, applies transorms to it,
    computes skecth from the given image,
    combines image and sketch and saves them as one image to the folder
    arguments:
        -img: image which was read with cv2,
        -dataset_path: path to store the dataset,
        -index: index of the given image, used to rename images by their indexes,
        -size: size of image after applying resize
        -crop_size: size of image after applying center crop
        -canny_threshold_1: threshold_1 used in canny edge detection implementation of opencv,
        -canny_threshold_2: threshold_2 used in canny edge detection implementation of opencv
    """
    img = transform(img, size=size, crop_size=crop_size)
    sketch = 255 - cv2.Canny(img, canny_threshold_1, canny_threshold_2)
    sketch = np.stack((sketch,)*3, axis=-1)   # make BW image have 3 color channels
    img_combined = np.concatenate([img, sketch], 1)
    write_combined(dataset_path, index, img_combined)

def save_sketch_only(img, 
                     dataset_path: str, 
                     index: int, 
                     size: tuple,
                     crop_size: tuple,
                     canny_threshold_1: int,
                     canny_threshold_2: int):
    """
    This function gets an image, applies transorms to it,
    computes skecth from the given image,
    combines sketch two times with itself and saves them as one image to the folder
    arguments:
        -img: image which was read with cv2,
        -dataset_path: path to store the dataset,
        -index: index of the given image, used to rename images by their indexes,
        -size: size of image after applying resize
        -crop_size: size of image after applying center crop
        -canny_threshold_1: threshold_1 used in canny edge detection implementation of opencv,
        -canny_threshold_2: threshold_2 used in canny edge detection implementation of opencv
    """
    img = transform(img, size=size, crop_size=crop_size)
    sketch = 255 - cv2.Canny(img, canny_threshold_1, canny_threshold_2)
    sketch = np.stack((sketch,)*3, axis=-1)   # make BW image have 3 color channels
    img_combined = np.concatenate([sketch, sketch], 1)
    write_combined(dataset_path, index, img_combined)
    
def load_images_from_folder(folder):
    return [f'{folder}/{img_name}' for img_name in os.listdir(folder)]

'''
  Takes paths to the folders with images and masks and returns lists 
  with all images and all masks paths one by one where masks[0] 
  corresponds to the imgs[0] and so on.
'''
def get_imgs_and_masks(imgs_folder, masks_folder):
    return [load_images_from_folder(imgs_folder), load_images_from_folder(masks_folder)]

def create_empty_dir(dir_name):
    if not os.path.exists(dir_name):
        os.makedirs(dir_name)
    else:
        shutil.rmtree(dir_path)
        os.makedirs(dir_name)

def split_dataset_and_write_on_disk(dataset: DatasetWithMasks, 
                                    train_size: float,
                                    img_size: tuple,
                                    crop_size: tuple,
                                    canny_threshold_1: int,
                                    canny_threshold_2: int):
    
    number_of_train_imgs = int(len(dataset) * train_size)
    
    train_dir = 'combined/train/'
    for i in range(number_of_train_imgs):
        save_img_and_masks(
            img=dataset[i],
            dataset_path=train_dir, 
            index=i, 
            size=img_size,
            crop_size=crop_size,
            canny_threshold_1=canny_threshold_1,
            canny_threshold_2=canny_threshold_2
        )
    
    test_dir = 'combined/test/'
    for i in range(number_of_train_imgs, len(dataset)):
        save_sketch_only(
            img=dataset[i],
            dataset_path=test_dir, 
            index=i-number_of_train_imgs, 
            size=img_size,
            crop_size=crop_size,
            canny_threshold_1=canny_threshold_1,
            canny_threshold_2=canny_threshold_2
        )
        
    return 'combined'

def preprocess_data(imgs: [str], masks: [str]):
    '''
        Takes list with all images paths and their masks paths, removed background
        on the images and applies canny edge detection to obtain sketch. Saves
        images without background and their sketches to the disk in the train/test folders.
        
        Returns the folder in which the combined images were saved in train/test folders.
    '''
    dataset = DatasetWithMasks(imgs, masks)
    
    # We can set different canny thresholds or sizes here
    return split_dataset_and_write_on_disk(
        dataset=dataset,
        train_size=0.8,
        img_size=(256,256),
        crop_size=(256,256),
        canny_threshold_1=80,
        canny_threshold_2=120
    )
```
